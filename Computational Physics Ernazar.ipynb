### Create factorial as a function
Don't need to define factorial program everywhere we need it. Define it once in function and call it wherever you need


```python
def factorial(n):
    f = 1
    for i in range(1, n+1):
        f = f*i
    #print(f) #to print the result
    return f #won't print, but will return the result to somewhere
```

```python
factorial(6)
```

```python
f4 = factorial(4)
f6 = factorial(6)#function returns the result into the variable f4
print(f4, f6)
```

## Program that takes polar coordinates as input to give cartesian coordinates as output 

```python
from math import sin, cos
```

```python
def polar2cartesian(r, theta):
    x = r*cos(theta)
    y = r*sin(theta)
    return x, y
```

```python
r0 = 10
theta0 = 0.1

x0, y0 = polar2cartesian(r0, theta0)
print(y0)
```

# Root finding and bracketing method


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def f(x):
   return  3.0*x**3 + x-3
```

```python
x = np.arange(0.5, 1.5, 0.1)
plt.plot(x, f(x))
plt.grid()
plt.show()
```

## Bisection algorithm

```python
def bisect(xl, xu, es, imax):
    #es: the precision to which I want the roots
    #imax: maximum number of iteratons
    xr=xu
    x=[]; e=[]; it=[]
    #keeps track of root, error, and iteration number
    for i in range(imax):
        xrold = xr
        xr = 0.5*(xl+xu)
        if(xr != 0.0):
            ea = abs((xr-xrold)/xr)*100
        it.append(i); x.append(xr); e.append(ea)
        test = f(xl)*f(xr)
        if(test<0.0):
            xu = xr
        elif(test>0.0):
            xl = xr
        else:
            ea = 0.0
        if(ea<es):
            break
    return it, x, e
        
```

## False positioning method

```python
def falsep(xl, xu, es, imax):
    xr = xu
    x=[]; e=[]; it=[]
    for i in range(imax):
        xrold = xr
        xr = xu - f(xu)*(xl-xu)/(f(xl)-f(xu))
        if(xr != 0.0):
            ea = abs((xr-xrold)/xr)*100
        it.append(i); x.append(xr); e.append(ea)
        test = f(xl)*f(xr)
        if(test<0.0):
            xu= xr
        elif(test>0.0):
            xl= xr
        else:
            ea = 0.0
        if(ea<es):
            break
    return it, x, e
```

```python
i1, x1, e1 = bisect(0.5, 1.5, 1e-12, 100)
i2, x2, e2 = falsep(0.5, 1.5, 1e-12, 100)

plt.plot(i1, e1, label = "Bisec");
plt.plot(i2, e2, label = "Falsep");

plt.legend()
plt.xlabel("Iteration number");
plt.ylabel("Error")
plt.yscale('log')
plt.show()
```
